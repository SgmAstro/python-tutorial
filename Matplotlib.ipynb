In this tutorial, we are going to better understand how to use matplotlib for data visualisation. This is one of the areas where Python shines.

To begin, let us import matplotlib.pyplot, as well as numpy:

```python
import matplotlib.pyplot as plt
import numpy as np
```

Next, let's use the plot function. To begin, this is how we plot 10 random numbers.

```python
plt.plot(np.random.rand(10))
```

We have a graph! That said, this graph is rather useless; right now the y-axis corresponds to the random number, and the x-axis is the order it is generated. Let's plot something more useful!

```python
x = [1, 2, 4, 7, 10, 11]
y = [2, 4, 9, 8, 17, 12]

plt.plot(x,y)
plt.title('Data from Experiment X')
plt.xlabel('time (secs)')
plt.ylabel('y value')
plt.show()
```

Here we go! Let's go through this a bit more slowly. plt.plot works as follows:

plt.plot(x,y)

where x and y are typically arrays of values.

It is also possible to have

plt.plot(y)

where x is taken to be the array index from 0 to N-1. Indeed, this is what happened when we plotted those random numbers above. 

In addition, we made the graph nicer by adding a title, an x-label, and a y-label.

Finally, we used the command plt.show(), which tells the graph to display itself at that point in the code. If you're astute, you may notice that this wasn't used earlier, and the graph of random numbers still displayed. Confused as to why this command is even necessary? Look at the following examples:


```python
x2 = [1,3,5,7,9,11]
y2 = [11,8,7,6,4,1]

plt.plot(x,y)
plt.plot(x2,y2)
```

```python
plt.plot(x,y)
plt.show()
plt.plot(x2,y2)
plt.show()
```

The above examples show that unless plt.show() is invoked, graphs may be drawn on the same plot. Now, let's move on to plotting functions:

```python
x = np.linspace(0,10,200)
y = np.cos(x)
y2 = np.cos(x**2)
plt.plot(x, y, label=r'$\cos(x)$')
plt.plot(x, y2, label=r'$\cos(x^2)$')
plt.title('Trig functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

Here we have given the graph a legend to distinguish between the two graphs, and added a grid. We can also play around with the lines of a graph. First, let us consider discrete points:

```python
plt.plot(x, y, 'o', markersize=2, color='r');
```

In the above code, the 'o' indicates that a circle marker is desired. Other useful markers include '.' (point), '+' (plus), and 'x' (cross).

The ability to specify the need for discrete points is highly desireable, especially if you want to plot a scattergraph without all these joined up lines. In fact, using 'o', you can change the random numbers graph at the very beginning to a scattergraph. Try it!

Also play around with the markersize and color parameters. You can review more parameters here: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html

```python
plt.plot(x, y, '--', linewidth=2, color='c');
```

There are additionally line styles that are useful for continuous line graphs, as seen above. Examples include '-' (solid), '--' (dashed), '-.' (dash-dot), and ':' (dotted).

Now we have a relatively strong understanding of matplotlib. And yet, there is still more. Let us look at another slightly different way of producing plots that will be slightly more useful for certain uses (and also useful when trying to understand Stack Overflow solutions).

```python
x = np.arange(0.0, 2.0, 0.01)
y = 1 + np.sin(2 * np.pi * x)

fig, ax = plt.subplots()
ax.plot(x, y)

ax.set(xlabel='time (s)', ylabel='value', title='Graph')
ax.grid()

plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), color='tab:blue', marker='o')
plt.plot(t2, f(t2), color='black')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), color='tab:orange', linestyle='--')
plt.show()
```

Finally, let's look at some of the other useful graph types that you may need to create:

```python
x = np.linspace(-5, 5)
y = np.exp(x**3)
plt.semilogy(x, y);
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)

plt.xlabel('IQ')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
# This will put a text fragment at the position given:
plt.text(55, .027, r'$\mu=100,\ \sigma=15$', fontsize=14)
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet,
        linewidth=0, antialiased=False)
ax.set_zlim3d(-1.01, 1.01);
```
